```python
# While Loops

name = input("Enter your name: ")

while name == "":
  name = input("Enter your name: ")

age = int(input("Enter your age: "))

while age < 0:
  age = int(input("Enter your age: "))

print(f"Hello {name}")
print(f"You are {age} years old")
```

```python
# List[] = mutable, most flexible

cars = ["bmw", "audi", "mustang", "mercedes"]
print(cars)

cars[0] = "ferrari"
print(cars)

print(cars[0])
print(cars[1])
print(cars[2])
print(cars[3])

#Adding to a list at the end
cars.append("toyota")
print(cars)

#So, this line of code effectively adds "honda" to the beginning of your cars list, shifting
#all existing elements one position to the right.
cars.insert(0, "honda")


#removing element
cars.remove("honda")
print(cars)

#We can pop meaning we can remove element at a given index
cars.pop(3)
print(cars)

for car in cars:
  print(car,end = " ")

cars.clear();

print()

print(cars)

```

```python
# Sets{} = mutable(add/remove) but we cannot replace any element with in it, unordered, We cannot access
#          by index
#          No duplicates , best for membership testing


countries = {"Pakistan","Iran","Turkey","Saudia Arabia","Qatar","Iraq","Iran"}

countries.add("Afghanistan")
print(countries)

countries.remove("Qatar")

# I am going to print the elements of my set,the sets are unordered the order is going to change each time
# during its first run we will have different order , during 2nd and other time we print , it will have
# different order

for country in countries:
  print(country, end = " ")

print()

# Sets are good for membership testing, checking to see if there is a given value within your set

if "Pakistan" in countries:
  print("Pakistant is in the set")
else:
  print("Pakistan is not in the set")


# We even could accept user input

country = input("Enter country name: ")
if country in countries:
  print(f"{country} is in the set")
else:
  print(f"{country} is not in the set")
```
